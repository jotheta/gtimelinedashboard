```python
import json
import pandas as pd
import os
import plotly.express as px

years = os.listdir(PATH)

list_of_activities = []

for year in years:
    if year != '.DS_Store':
        
        months = os.listdir(f'{PATH}/{year}')
        
        for month in months: 
            if month != '.DS_Store':
                
                final_path = f'{PATH}/{year}/{month}'
                
                with open(final_path) as json_file:
                    file = json.load(json_file)
                    
                    list_of_activities.extend(file['timelineObjects'])
```

```python
location_dict = {}
coordinate_scaler = 1e7

for point in list_of_activities:
    if list(point.keys())[0] == 'placeVisit':
        
        placeId = point['placeVisit']['location']['placeId']
        
        if placeId not in location_dict.keys():
            
            if all(check in point['placeVisit']['location'].keys() for check in ('longitudeE7', 'latitudeE7', 'name')):
                
                lon = point['placeVisit']['location']['longitudeE7'] / coordinate_scaler
                lat = point['placeVisit']['location']['latitudeE7'] / coordinate_scaler
                name = point['placeVisit']['location']['name']
                visits = 1
            
            location_dict[placeId] = {'lon': lon, 'lat': lat, 'name': name, 'visits': visits}
            
        else:
            location_dict[placeId]['visits'] += 1
            


location_df = pd.DataFrame.from_dict(location_dict, orient='index').reset_index()
location_df.rename(columns={'index':'placeId'}, inplace=True)
        
filtered_location_df = location_df.loc[location_df['visits'] <= 10000]
```

```python
fig = px.scatter_mapbox(filtered_location_df,
                        lat='lat',
                        lon='lon',
                        size='visits',
                        color='visits',
                        hover_name='name')

fig.update_layout(mapbox_style='stamen-terrain')

fig.show()
```

```python
activity_dict = {}
coordinate_scaler = 1e7

for i, point in enumerate(list_of_activities):
    
    if list(point.keys())[0] == 'activitySegment':
        
        print(list(point.keys())[0])
        print(point['activitySegment']['startLocation'])
        
        if 'longitudeE7' in point['activitySegment']['startLocation']:
            s_lon = point['activitySegment']['startLocation']['longitudeE7'] / coordinate_scaler
            s_lat = point['activitySegment']['startLocation']['latitudeE7'] / coordinate_scaler
            e_lon = point['activitySegment']['endLocation']['longitudeE7'] / coordinate_scaler
            e_lat = point['activitySegment']['endLocation']['latitudeE7'] / coordinate_scaler
            
        activity_dict[i] = {'s_lon': s_lon, 's_lat': s_lat, 'e_lon': e_lon, 'e_lat': e_lat}
            


activity_df = pd.DataFrame.from_dict(activity_dict, orient='index').reset_index()
```
